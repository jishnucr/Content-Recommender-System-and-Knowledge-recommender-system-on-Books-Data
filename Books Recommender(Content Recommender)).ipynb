```python
def build_chart(df2,percentile = 0.08):
    print("input Preferred Categories")
    categories = str(input())
    print("input year")
    year = int(input())
    books = df2.copy()
    books = books[(books["categories"]==categories) & (books["published_year"]==year)]
    C = books["average_rating"].mean()
    m = books["ratings_count"].quantile(percentile)
    q_books = books.copy().loc[books["ratings_count"]>=m]
    q_books["score"] = q_books.apply(lambda x: (x["ratings_count"]/(x["ratings_count"]+m)* x["average_rating"])+(m/(m+x["ratings_count"])*C),axis=1)
    q_books = q_books.sort_values(by='score',ascending=False)
    return q_books
```

```python
build_chart(df2)
```

### now Lets build a Content Based recommender Sysytem

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df2["description"])
```

```python
tfidf_dense = tfidf_matrix.toarray()
```

```python
tfidf_dense
```

## Computation

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_dense,tfidf_dense)
cosine_sim
```

## Building the Recommender System

```python
indices = pd.Series(df2.index,index=df2["title"]).drop_duplicates()
```

```python
indices
```

```python
def content_recommender(title,cosine_sim=cosine_sim,df2=df2,indices=indices):
    idx=indices[title]
    sim_score = list(enumerate(cosine_sim[idx]))
    sim_score = sorted(sim_score,key=lambda x: x[1],reverse=True)
    sim_score = sim_score[1:11]
    books_indices = [1[0] for i in sim_score]
    return df2['title'].iloc[books_indices]
```

```python
content_recommender('The Four Loves')
```
